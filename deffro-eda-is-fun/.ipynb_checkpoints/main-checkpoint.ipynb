```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
print("hello")
```

```python
train = pd.read_csv('../data/train_V2.csv')
train.info()
```

```python
print("The average person kills: {:.4f} players, 99% of people have {} kills or less, while the most kills ever recorded is {}."
      .format(train['kills'].mean(), train['kills'].quantile(0.99), train['kills'].max()))
```

```python
data = train.copy()
data.loc[data['kills'] > data['kills'].quantile(0.99)] = '8+'
plt.figure(figsize=(15, 10))
sns.countplot(data['kills'].astype('str').sort_values())
plt.title("kill count", fontsize=15)
plt.show()
```

```python
# Do those players who cannot make a single kill cause damage to others?
data = train.copy()
data = data[data['kills']==0]
plt.figure(figsize=(15, 10))
plt.title("Damage Dealt by 0-killers", fontsize=15)
sns.distplot(data['damageDealt']) # hist for real-value x?
plt.show()
```

```python
# so 0-kills player mostly cannot cause damage to others
# Let's investigate the exceptions
print("{} players ({:.4f}%) have won without a single kill".format(
    len(data[data['winPlacePerc']==1]), 100*len(data[data['winPlacePerc']==1]) / len(train) 
))

data1 = train[train['damageDealt']==0].copy()
print("{} players ({:.4f}%) have won without dealing damage!".format(
    len(data1[data1['winPlacePerc']==1]), 100*len(data1[data1['winPlacePerc']==1]) / len(train) 
))
```

```python
# plot win placement percentage vs kills
sns.jointplot(x="winPlacePerc", y="kills", data=train, height=10, ratio=3, color="#6AB27B")
plt.show()
```

```python
# from above we can obtain that killing has a correlation with winning. 
# We group players on kills (0 kills, 1-2 kills, 3-5 kills, 6-10 kills and 10+ kills)
kills = train.copy()
kills['killsCategories'] = pd.cut(kills['kills'], [-1, 0, 2, 5, 10, 72], 
                                  labels=['0_kills', '1-2_kills', '3-5_kills', '6-10_kills', '10+_kills'])
plt.figure(figsize=(15, 8))
sns.boxplot(x="killsCategories", y="winPlacePerc", data=kills)
plt.show()
```

```python
print("The average person walks for {:.1f}m, 99% of people have walked {}m or less, while the marathoner champion walked for {}m."
      .format(train['walkDistance'].mean(), train['walkDistance'].quantile(0.99), train['walkDistance'].max()))
```

```python
# plot walk distance distribution
data = train.copy()
data = data[data['walkDistance'] < train['walkDistance'].quantile(0.99)]
plt.figure(figsize=(15, 10))
plt.title('Walking Distance Distribution', fontsize=15)
sns.distplot(data['walkDistance'])
plt.show()
```

```python
print("{} players ({:.4f}%) walked 0 meters. This means that they die before even taking a step or they are afk."
      .format(len(data[data['walkDistance']==0]), 100*len(data[data['walkDistance']==0])/len(train)))
```
